## REFSYS

````python

    ###  REFSYS ------ Specifies the coordinate system of integration  
    ###     key        and the number of arc iterations 
    ###   
    ###  Columns   Format   Description   
    ###  -------   ------   ----------------------------
    ###   1-6       A6     REFSYS  (CARD name)
    ###                         
    ###   7-8       I2     Maximum number of arc iterations during the first global iteration.
    ###                               
    ###   9         I1     Minimum number of arc iterations during the first global iteration.     
    ###                           
    ###   10        I1     Maximum number of arc iterations during global iterations after the first.      
    ###                          
    ###   11        I1    Coordinate reference system
    ###                        0 True of reference date and time
    ###                        1 Mean of J2000 [see note 1]
    ###                           
    ###   21-26     I6    Year ,month ,day of reference date (YYMMDD ).  
    ###                           
    ###   27-30     I4    Hour ,minute of reference time (HHMM).
    ###                           
    ###   31-40    D10.8   Seconds of reference time (SS.SSSSSSS ).     
    ###                           
    ##### NOTE: [1] The DE200 Ephemeris is needed when the J2000 reference system is selected. 
    ###             No reference date is needed in cols. 21-40 if J2000 is selected.


CARD_refsys             = 'REFSYS'   #  1-6       A6
max_arciters_global1    = 19         #  7-8       I2
min_arciters_global1    = 3          #  9         I1
max_arciters_global_2on = 3          #  10        I1
coordinate_system       = 1          #  11        I1
YYMMDD                  = 181109     #  21-26     I6
HHMM                    = 0000       #  27-30     I4
SS_SSSSSSS              = 00.0000000 #  31-40    D10.8


cols__1_6   = str(     CARD_refsys              ).rjust(6,' ') #A6
cols__7_8   = str( int(max_arciters_global1)    ).rjust(2,' ') #I2
cols__9     = str( int(min_arciters_global1)    ).rjust(1,' ') #I1
cols__10    = str( int(max_arciters_global_2on) ).rjust(1,' ') #I1
cols__11    = str( int(coordinate_system)       ).rjust(1,' ') #I1
cols__13_20 = '         '
cols__21_26 = f'{YYMMDD}'.rjust(6,' ')  #I6
cols__27_30 = f'{HHMM:04}'.rjust(4,' ')  #I4
cols__31_40 = f'{SS_SSSSSSS:010.7f}'.rjust(10,' ') #D10.8

                  #The first number is the total number of digits,
                  #   including decimal point. 
                  #   Both decimal points (.) and minus signs (-) 
                  #   are included in the count.

refsys_line = cols__1_6   + cols__7_8   + \
              cols__9     + cols__10    + cols__11 + \
              cols__13_20 + cols__21_26 + \
              cols__27_30 + cols__31_40 

# print(line_nos1,sep='')
# print(line_nos2,sep='')
# print(line_nos3,sep='')

# print(refsys_line,sep='')
````

## SATPAR

````python

    ###  SATPAR ------Specifies spacecraft identification number (Satellite ID), 
    ###     key       cross sectional area and mass. Also specifies maximum 
    ###               degree of gravitational coefficients to be considered
    ###               significant for this spacecraft.
    ###   
    ###  Columns   Format   Description   
    ###  -------   ------   ----------------------------
    ###   1-6       A6     SATPAR  (CARD name)
    ###                         
    ###   7         I1    Choose Cross Sectional Area and Mass
    ###                       0 - Area and Mass defined on this card is used
    ###                       1 - Obtain same from S/C parameters file.
    ###   8         I1  
    ###                       0 - No tables of S/C cross section or 
    ###                           solar radiation to be obtained
    ###                           from S/C parameter file.
    ###                       1 - Only drag cross section tables
    ###                           used from S/C parameter file.
    ###                       2 - Only solar radiation tables used
    ###                           from S/C parameters file.
    ###                       3 - Both drag cross section and solar
    ###                           radiation tables used from S/C parameters file.
    ###                         
    ###   9         I1    Twelve parameter state combination identifier.
    ###                     If this index is present there must be one more SATPAR card
    ###                     in the unit 5 setup to form the pair. For the fisrt satellite
    ###                     the index must be 1 and for the second 2. For the satellites
    ###                     forming a pair with these indices on col. 9, the following 
    ###                     ELEMS1 and ELEMS2 cards contain 6 (six) each of the twelve
    ###                     state vector combination parameters. 
    ###                     The first SATPAR with the index 1 is followed by ELEMS cards
    ###                     featuring the mid-point parameters , and the second SATPAR
    ###                     is followed by ELEMS cards featuring baseline parameters (Ref. NOTE [2]).
    ###                     Any following VARCOV cards for these two satellites feature
    ###                     the sigmas of the 12 state combination parameters. 
    ###                           
    ###   10-11     I2    Spacecraft attitude control algorithm
    ###                      0 - Cannonball(no attitude control)  2 - TOPEX/Poseidon
    ###                      3 - SPOT -2                          4 - GPS 
    ###                      5 - ERS -1                           6 - Mars Observer Mapping Phase
    ###                      7 - Mars Observer Cruise Phase       8 - TDRSS 
    ###                      9 - Magnetically stabilized S/C      10 - GFO
    ###                      11 - TRMM                            12 - EUVE 
    ###                      13 - ICESAT                          14 - ENVISAT
    ###                      15 - CRYOSAT -2                      16 - Leading GRAIL satellite
    ###                      17 - Trailing GRAIL satellite        18 - HY2A 
    ###                      19 - SARAL                           20 - SENTINEL
    ###                           
    ###   12        I1    Requests external attitude for this satellite:
    ###                      0 - No external attitude 
    ###                      9 - External attitude 
    ###                          (See Volume 5 for a description 
    ###                             of the external attitude file)
    ###                           
    ###   13        I1    Local gravity application switch
    ###                      0 - Consider gravity anomaly, surface density
    ###                          and geoid height local gravity terms
    ###                          to be significant for this spacecraft.
    ###                      1 - Consider local gravity terms to be 
    ###                          insignificant for this
    ###
    ###   14        I1    Requests external thermal acceleration for this satellite:
    ###                      0 - No external thermal acceleration 
    ###                      1 - External thermal acceleration 
    ###                          (See Volume 5 for a description 
    ###                             of the external acceleration file)    
    ###                      
    ###   15-17     I3    Maximum degree of gravitational coefficients 
    ###                     Max. of Model to be considered significant for this S/C.
    ###                           
    ###   18-24     I7    Satellite ID (satellite ID is required ).
    ###                           
    ###   25-44   D20.8   S/C cross sectional area.   (m^2)
    ###                           
    ###   45-59   D15.3   S/C mass.   (Kg)
    ###### 
    ### UNITS: KM =Kilometers; M =Meters ; S =Seconds      ; Kg =Kilograms 
    ###        DEG=Degrees ; RAD=Radians ; AS =Arc seconds ; MAS=Milli-arc seconds
    ###
    ### NOTE: [1] If more than one satellite is used in an arc than the 
    ###           four lines SATPAR , EPOCH , ELEMS1 ELEMS2 - are repeated for each satellite.


CARD_satpar             = 'SATPAR'   
SC_param_A_m       = 0      # Use S/C area and mass from this card
SC_param_DragSRP   = 0      # No S/C param file used        
TwelveParam        = 0      # not used
AttitudeControl    = 0      # cannonball          
ExternalAttitude   = 9      # use EXTATT file
LocalGravitySwitch = 0      # Gravity is considered significant
Ext_ThermalAccel   = 0      # No external thermal acceleration
MaxDeg_GravCoeffs  = 0      # default-- Max. of Model
# SATID = SATID
SC_CrossSec_Area   =  9999      #    (m^2)
SC_Mass            =  4.933     #   (Kg)


#### Put each input option into the correct format  
cols__1_6   = f'{CARD_satpar        }'.rjust(6,' ') #A6
cols__7     = f'{SC_param_A_m       }'.rjust(1,' ') #I1
cols__8     = f'{SC_param_DragSRP   }'.rjust(1,' ') #I1
cols__9     = f'{TwelveParam        }'.rjust(1,' ') #I1
cols__10_11 = f'{AttitudeControl    }'.rjust(2,' ') #I2
cols__12    = f'{ExternalAttitude   }'.rjust(1,' ') #I1
cols__13    = f'{LocalGravitySwitch }'.rjust(1,' ') #I1
cols__14    = f'{Ext_ThermalAccel   }'.rjust(1,' ') #I1
cols__15_17 = f'{MaxDeg_GravCoeffs  }'.rjust(3,' ') #I3
cols__18_24 = f'{SATID              }'.rjust(7,' ') #I7
cols__25_44 = f'{SC_CrossSec_Area:20.8f}'.rjust(20,' ')#D20.8
cols__45_59 = f'{SC_Mass         :15.3f}'.rjust(15,' ')#D15.3


                  #The first number is the total number of digits,
                  #   including decimal point. 
                  #   Both decimal points (.) and minus signs (-) 
                  #   are included in the count.

satpar_line = cols__1_6       \
               + cols__7      \
               + cols__8      \
               + cols__9      \
               + cols__10_11  \
               + cols__12     \
               + cols__13     \
               + cols__14     \
               + cols__15_17  \
               + cols__18_24  \
               + cols__25_44  \
               + cols__45_59

print(line_nos1,sep='')
print(line_nos2,sep='')
print(line_nos3,sep='')

print('SATPAR   139     1807001          9.53000000       1514.000 ')
print(satpar_line,sep='')
````

## Panel Model Info

````python
# Panel Model Transcribed from Eric Sutton's getareanrm.m
"""
This function establishes the 
% macromodel plate areas (m^2) and
% plate normal vectors in spacecraft body-fixed coordinates
% data refers to Spire/LEMUR2_3.3_defeatured.SLDPRT geometry file and body-fixed coordinate system
"""
import numpy as np
    

nrm_sbf = {} #np.zeros( 8 )    # plate normal vectors in spacecraft body-fixed coordinates
A       = {} #np.zeros( 8 )    # plate areas (m^2)
plt_M   = {} #np.zeros( 8 )    # plate materials
bwspec  = {} #np.zeros( 8 )    #
bwdiff  = {} #np.zeros( 8 )    #

Au = 196.96655  # gold atomic mass
#SiO2 = 60.0843
SiO2 = 20.03    # silicon dioxide atomic mass
Al = 26.981538  # aluminum atomic mass

# nrm_sbf = np.zeros( (8,3) )

### -------
### PANEL 1
### -------
###         **** BODY: Front (FRO Ant.) panel +X
nrm_sbf[1] = np.array([ 1, 0, 0])
A[1] = 0.1053314*0.3388226;
bwspec[1] = 0.2
bwdiff[1] = 0.4
plt_M[1]  = Al

### ----------
### PANEL 2
### ----------
###         **** BODY: Port -Y
nrm_sbf[2] = np.array([ 0, -1, 0])
A[2]       = (0.105186*0.3388226) + (3*0.005334*0.074549)
bwspec[2]  = 0.2
bwdiff[2]  = 0.4
plt_M[2]   = Al

### -------
### PANEL 3
### -------
###         **** SOLAR PANELS: "Front" panels (only half covered w/ solar panels, so split into halves (1/2 SiO2 and 1/2 Al) if using SRP model
nrm_sbf[3] = (np.sqrt(2)/2) * np.array([ 1,-1, 0])
A[3]       = 2*0.150*0.3222
bwspec[3]  = 0.05
bwdiff[3]  = 0.3
plt_M[3]   = SiO2

### -------
### PANEL 4
### -------
###         **** BODY: Aft panel -X (mostly covered in solar panels (as of rev. 3.3 & prior to rev. 3.4)
nrm_sbf[4] = np.array([-1, 0, 0])
A[4]       = 0.1053314*0.3388226
bwspec[4]  = 0.05
bwdiff[4]  = 0.3
plt_M[4]   = SiO2

### -------
### PANEL 5
### -------
###         **** BODY: Star +Y (mostly covered in solar panels)
nrm_sbf[5] = np.array([ 0, 1, 0])
A[5]       = (0.105186*0.3388226) + (3*0.005334*0.074549)
bwspec[5]  = 0.05
bwdiff[5]  = 0.3
plt_M[5]   = SiO2


### -------
### PANEL 6
### -------
###         **** SOLAR PANELS: "Back/Aft" panels (mostly covered w/ solar panels)
nrm_sbf[6] = (np.sqrt(2)/2)* np.array([-1, 1, 0])
A[6]       = 2*0.150*0.3222
bwspec[6]  = 0.05
bwdiff[6]  = 0.3
plt_M[6]   = SiO2

### -------
### PANEL 7
### -------
###         **** BODY: Zenith (POD ant.) -Z
nrm_sbf[7] = np.array([ 0, 0, -1])
A[7]       = 0.105186*0.1053314
bwspec[7]  = 0.2
bwdiff[7]  = 0.4
plt_M[7]   = Al


### -------
### PANEL 8
### -------
###         **** BODY: Nadir +Z
nrm_sbf[8] = np.array([0, 0, 1])
A[8]       = 0.105186*0.1053314
bwspec[8]  = 0.2
bwdiff[8]  = 0.4
plt_M[8]   = Al

# # for i = 1:length(nrm_sbf),
# #     nrm_sbf(:,i) = nrm_sbf(:,i)./norm( nrm_sbf(:,i) );
# # end

for ipanel, num  in enumerate([1,2,3,4,5,6,7,8]):

    print("PANEL", num,'-----------------')
    print('   NormVec ',nrm_sbf[num])
    print('   Area    ',A[num])
    print('   bwspec  ',bwspec[num])
    print('   bwdiff  ',bwdiff[num])
    print('   Material',plt_M[num])
    print()
````
