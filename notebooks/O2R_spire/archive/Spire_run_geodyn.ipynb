```python
dates       = Obj.__dict__['Trajectory_orbfil']['2018.313']['data_record']['Date_UTC']
#
EastLon     = Obj.__dict__['Trajectory_orbfil']['2018.313']['data_record']['Satellite East Longitude']
GeodeticLat = Obj.__dict__['Trajectory_orbfil']['2018.313']['data_record']['Satellite Geodetic Latitude']
Height      = Obj.__dict__['Trajectory_orbfil']['2018.313']['data_record']['Satellite Height']


```

### Load Spire OD for compare


```python
from time_systems import time_gps_to_tdt
from quaternions  import load_attitude_spire

startEpoch  = "2018-11-08 23:00:00"  
stopEpoch   = "2018-11-10 01:00:00" 
interval    = 10  # seconds
    

path_attitude     = "/data/data_geodyn/inputs/spire83/data_Spire/attitude/20180923_20181209_TaskOrder3Data/"
filename          = "leoAtt_2018-09-23T00-12-00Z.9999999.083.log"
file__AttitudeLog = path_attitude+filename

#### 1. Load the attitude data that corresponds to the entire timeperiod
SpireDF = load_attitude_spire(file__AttitudeLog,
                                  startEpoch,
                                  stopEpoch)


#### 2. Convert from GPS time to TDT time
SpireDF['tim (tdt)'] = [time_gps_to_tdt(tim, leap_sec=37) 
                                 for tim in SpireDF['tim (gps)'] 
                                        ]
```

```python
SpireDF
```

### Plot

```python
x_spire = []
y_spire = []
z_spire = []

for i,val in enumerate(SpireDF['pos (eci)']):
#     print(val[0])
    x_spire.append(val[0]*1e3)
    y_spire.append(val[1]*1e3)
    z_spire.append(val[2]*1e3)
```

```python
import plotly.graph_objects as go
from plotly.offline import plot, iplot
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.io as pio   ### Allows you to save plotly figs


fig  = make_subplots(
    rows=1, cols=1,
    subplot_titles=([' ',' ']),
    shared_xaxes=False,
    vertical_spacing = 0.1,
            )

################S#######################################
font_dict=dict(family='Arial',size=14,color='black')
#######################################################

col_r  =  "#ff7f0e"  # 'tab:orange'
col_b  =  "#1f77b4"  # 'tab:blue'


fig.add_trace(go.Scattergl(x=dates,
                           y=Obj.__dict__['Trajectory_orbfil']['2018.313']['data_record']['Satellite Inertial X coordinate'],
                           mode='markers+lines',
                           marker = dict(size=4, color=col_b),
                           opacity=1,
                           showlegend=False), row=1, col=1)
fig.add_trace(go.Scattergl(x=dates,
                           y=Obj.__dict__['Trajectory_orbfil']['2018.313']['data_record']['Satellite Inertial Y coordinate'],
                           mode='markers+lines',
                           marker = dict(size=4, color=col_b),
                           opacity=1,
                           showlegend=False), row=1, col=1)
fig.add_trace(go.Scattergl(x=dates,
                           y=Obj.__dict__['Trajectory_orbfil']['2018.313']['data_record']['Satellite Inertial Z coordinate'],
                           mode='markers+lines',
                           marker = dict(size=4, color=col_b),
                           opacity=1,
                           showlegend=False), row=1, col=1)

fig.add_trace(go.Scattergl(x = SpireDF['tim (gps)'],
                           y =x_spire,
                           mode='markers+lines',
                           marker = dict(size=4, color=col_r),
                           opacity=1,
                           showlegend=False), row=1, col=1)
fig.add_trace(go.Scattergl(x = SpireDF['tim (gps)'],
                           y = y_spire,
                           mode='markers+lines',
                           marker = dict(size=4, color=col_r),
                           opacity=1,
                           showlegend=False), row=1, col=1)
fig.add_trace(go.Scattergl(x = SpireDF['tim (gps)'],
                           y = z_spire,
                           mode='markers+lines',
                           marker = dict(size=4, color=col_r),
                           opacity=1,
                           showlegend=False), row=1, col=1)

fig.update_yaxes(title_text="Inertial XYZ (m)",exponentformat= 'power',row=1, col=1)
fig.update_xaxes(title="Date", row=1, col=2)


fig.update_layout(title = '24-Hour Orbit Propagation of Spire #83',
                  autosize=True,
#                   autosize=False,    width=1000,    height=800,
                  legend= {'itemsizing': 'trace'},
                  font=font_dict,
                 )
fig.show(config=dict({
            'displayModeBar': False,
            'responsive': False,
            'staticPlot': False,
            'displaylogo': False,
            'showTips': False,
            }))
```

```python
import plotly.graph_objects as go
from plotly.offline import plot, iplot
from plotly.subplots import make_subplots
import plotly.express as px
import plotly.io as pio   ### Allows you to save plotly figs


fig  = make_subplots(
    rows=2, cols=1,
    subplot_titles=([' ','   ']),
    shared_xaxes=False,
    vertical_spacing = 0.1,
            )

################S#######################################
font_dict=dict(family='Arial',size=14,color='black')
#######################################################

col_r  =  "#ff7f0e"  # 'tab:orange'
col_b  =  "#1f77b4"  # 'tab:blue'


fig.add_trace(go.Scattergl(x=dates,
                           y=Height*1e-3 ,
                           mode='markers+lines',
                           marker = dict(size=4, color=col_b),
                           opacity=1,
                           showlegend=False), row=1, col=1)
fig.add_trace(go.Scattergl(x=GeodeticLat,
                           y=EastLon    ,
                           mode='markers+lines',
                           marker = dict(size=4, color=col_b),
                           opacity=1,
                           showlegend=False), row=2, col=1)

fig.update_yaxes(title_text="Altitude",row=1, col=1)
fig.update_xaxes(title="Date", row=1, col=2)

fig.update_yaxes(title_text="Geodetic Latitude",exponentformat= 'power',row=2, col=1)
fig.update_xaxes(title_text="East Longitude",exponentformat= 'power',row=2, col=1)


fig.update_layout(title = '24-Hour Orbit Propagation of Spire #83',
                  autosize=False,    width=1000,    height=800,
                  legend= {'itemsizing': 'trace'},
                  font=font_dict,
                 )
fig.show(config=dict({
            'displayModeBar': False,
            'responsive': False,
            'staticPlot': False,
            'displaylogo': False,
            'showTips': False,
            }))
```
