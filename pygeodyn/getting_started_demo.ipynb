# Example: How to Read GEODYN Ouput


```python
%load_ext autoreload
%autoreload 2

import sys  
sys.path.insert(0, '/data/geodyn_proj/pygeodyn/utils_pygeodyn/')
from pygeodyn_Read import pyGeodyn_Readers
```

### To input data from a single arc (based on the run parameters):

```python
#### first chnage the data path to be actual data
run_params['SpecialRun_name']  =  ''  


Obj_read_geodyn = pyGeodyn_Readers(run_params)
Obj_read_geodyn.getData_All()
```

```python
print(Obj_read_geodyn.__dict__.keys())
```

### To input data from all arcs:

```python
#### First modify the input parameters
run_params['arc']              =  [ '030914_2wk',
                                    '030928_2wk',
                                    '031012_2wk',
                                    '031026_2wk',
                                    '031109_2wk',
                                    '031123_2wk',
                                    '031207_2wk',
#                                     '031221_2wk',
                                  ]  

#### Re-initialize the reader object with the new parameters
Obj_read_geodyn = pyGeodyn_Readers(run_params)
Obj_read_geodyn.getData_All()
```
